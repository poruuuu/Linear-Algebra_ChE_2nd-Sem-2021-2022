###Representing Vectors
A vector could be thought of as a one-dimensional array in simple terms. A vector in Python refers to a one-dimensional array of sequences. It utilizes the items in the same way as a Python list operates. Let's say that you have vectors:

$$ A = 2\hat{x} + 3\hat{y} \\
B = 4\hat{x} - 2\hat{y}\\
C = 1ax + 7ay - 6az \\
D = 4\hat{i} - 3\hat{j} + 1\hat{k}$$


the equivalent of the above presented vector in matrix is as follows:

$$ A = \begin{bmatrix} 2 \\ 3\end{bmatrix} , B = \begin{bmatrix} 4 \\ -2\end{bmatrix} , C = \begin{bmatrix} 1 \\ 7 \\  -6 \end{bmatrix}, D = \begin{bmatrix} 4 \\ -3 \\ 1\end{bmatrix}
$$
$$ A = \begin{bmatrix} 2 & 3\end{bmatrix} , B = \begin{bmatrix} 4 & -2\end{bmatrix} , C = \begin{bmatrix} 1 & 7 & -6\end{bmatrix} , D = \begin{bmatrix} 4 & -3 & 1\end{bmatrix} 
$$

###Visualizaing Data

```python
import matplotlib.pyplot as plt
import matplotlib
%matplotlib inline
```

```python
H = [7, 61]
J = [-47, 2]
plt.scatter(H[0], H[1], label='H', c='yellow')
plt.scatter(J[0], J[1], label='J', c='red')
plt.grid()
plt.legend()
plt.show()
```

```python
H = np.array([7, 61])
J = np.array([-47, 2])
plt.title("Resutant Vector")
plt.xlim(-70, 70)
plt.ylim(-70, 70)
plt.quiver(0, 0, H[0], H[1], angles='xy', scale_units='xy', scale=1, color='green')
plt.quiver(H[0], H[1], J[0], J[1], angles='xy', scale_units='xy', scale=1, color='blue')
I = H + J
plt.quiver(0, 0, I[0], I[1], angles='xy', scale_units='xy', scale=1, color='violet')
plt.grid()
plt.show()
print(I)
Slope = I[1]/I[0]
print(Slope)
Angle = (np.arctan(Slope))*(180/np.pi)
print(Angle)
```
